```python
import pandas as pd

Generaciones = {
    "Baby Boomers": ["1946-1960", 10],
    "Generación X": ["1961-1980", 0],
    "Millennials": ["1981-1995", 70],
    "Generación Z": ["1996 en adelante", 20]}
```

```python
import matplotlib.pyplot as plt

df_Generaciones = pd.DataFrame.from_dict(Generaciones, orient='index', columns=['Rango de años', 'Porcentaje'])
print(df_Generaciones)

plt.barh(df_Generaciones.index, df_Generaciones['Porcentaje'])
plt.xlabel('Porcentaje')
plt.ylabel('Generación')
plt.title('Distribución de porcentajes por generación')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame(columns=['Categoría', 'Salario Diario Inicial', 'Salario Diario Intermedio - Bajo',
                             'Salario Diario Intermedio - Alto', 'Salario Diario Máximo', 'Personal a cargo',
                             'Tiempo mínimo entre ascensos', '% de personal en categoría', 'Nivel promedio de escolaridad'])

data = data.append({'Categoría': 'Avance - 30%',
                    'Salario Diario Inicial': 300.00,
                    'Salario Diario Intermedio - Bajo': 350.00,
                    'Salario Diario Intermedio - Alto': 400.00,
                    'Salario Diario Máximo': 500.00,
                    },ignore_index=True )

fig, ax = plt.subplots(figsize=(15, 7), dpi=100)
boxplot = data.boxplot(column=['Salario Diario Inicial', 'Salario Diario Intermedio - Bajo',
                               'Salario Diario Intermedio - Alto', 'Salario Diario Máximo'], 
                       labels=['Salario Diario Inicial', 'Salario Diario Intermedio - Bajo',
                               'Salario Diario Intermedio - Alto', 'Salario Diario Máximo'])

ax.set_ylim()
ax.set_title("Distribución de Salarios")
ax.set_xlabel("Categoría")
ax.set_ylabel("Salario Diario")
plt.show()



```
